#### EMPLOYEE SALES DATA ANALYSIS 

Import of required libraries 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Getting required datasets for analysis 

```python
fg= pd.read_csv(r"C:\Users\OPEYODE\Desktop\N-Tech\Salaries.csv")
fg
```

#### Data Cleaning: Checking dataset for missing values

```python
fg.isnull().sum()
```

#### Data Cleaning: Cleaning of the dataset by removing the "Notes" row as data is insufficient.

```python
cl= fg.drop(["Notes", "Status"],axis=1)
cl
```

#### Data Cleaning: Checking if above action was successful.

```python
cl.isnull().sum()
```

#### 1. Empoyee analysis: I filltered the dataset to call out the top five employees 

```python
top_5_employee_names = cl['EmployeeName'].value_counts().head(5)
top_5_employee_names
```

#### 2. Job analysis: Finding out how many unique job titles there are.

```python
unique_job_titles_count = cl['JobTitle'].nunique()
unique_job_titles_count
```

#### 3. Job analysis: Finding the total number of jobs contaning captain

```python
captain_job_titles_count = cl[cl['JobTitle'].str.contains('captain', case=False, na=False)].shape[0]
captain_job_titles_count
```

#### 4. Departmental Analysis: I filtered the dataset to focus on employees in the Fire Department.

```python
fire_dept_employees = cl[cl['JobTitle'].str.contains('fire department', case=False, na=False)]['EmployeeName']
print(fire_dept_employees)
```

#### 5. To understand the salary distribution, I calculated the minimum, maximum, and average BasePay. This helped me identify the lowest-paid and highest-paid employees and understand overall pay trends.

```python
#Find The Minimum,Maximum and Average BasePay

min_basepay = cl['BasePay'].min()
max_basepay = cl['BasePay'].max()
avg_basepay = cl['BasePay'].mean()

min_basepay, max_basepay, avg_basepay
```

#### 6. Data Cleaning: I replaced all 'Not provided' entries in the EmployeeName column with NaN values to treat missing names consistently.

```python
# Replace 'Not provided' in EmployeeName Column to Nan

cl['EmployeeName'] = cl['EmployeeName'].replace('Not provided', pd.NA)
cl
```

#### 7. Incomplete Data: I realized that some rows contained multiple missing values. I decided to drop rows with five or more missing values.

```python
#Drop the Rows Having 5 Missing Values:
cl_cleaned = cl.dropna(thresh=len(cl.columns) - 5)
cl_cleaned
```

#### 8. Identifying Specific Employees: Finding the actual job title of ALBERT PARDINI.

```python
job_title_albert = cl[cl['EmployeeName'] == 'ALBERT PARDINI']['JobTitle'].values[0]
print(job_title_albert)
```

#### 9. Identifying Specific Employees:  I found ALBERT PARDINI's job title and now calculated his total compensation, including benefits.

```python
#9.How Much ALBERT PARDINI Makes (include Benefits):

total_pay_benefits_albert = cl[cl['EmployeeName'] == 'ALBERT PARDINI']['TotalPayBenefits'].values[0]
print(total_pay_benefits_albert)
```

#### 10. Identifying Specific Employees: i identified the employee having the highest BasePay.

```python
#10.Display Name Of The Person Having The Highest BasePay:

highest_basepay = cl.loc[cl['BasePay'].idxmax()]['EmployeeName']
highest_basepay
```

#### 11. Analysis by Year: One of my goals was to find the average BasePay of employees for each year. This would help track pay trends over time.

```python
#11. Find The average Basepay of all Employees per Year

avg_basepay_per_year = cl.groupby('Year')['BasePay'].mean()
avg_basepay_per_year
```

#### 12. Analysis by Job Title: I calculated the average BasePay for all employees, grouped by their job titles. This provided insights into which roles were paid the most and least.

```python
#12. Find the Average BasePay of All Employees per Job Title:
avg_basepay_per_job = cl.groupby('JobTitle')['BasePay'].mean()
avg_basepay_per_job
```

#### 14. Common Job Titles: Finally, I identified the five most common job titles in the dataset.

```python
#14. Find the Five Most Common Job Titles:
common_job_titles = cl['JobTitle'].value_counts().head(5)
common_job_titles
```
